```python
import pandas as pd
import numpy as np
```

```python
list_dfs = []
for i in range(1,32):
    tmp_df =  pd.read_csv('Aemet2018-01-'+str(i).zfill(2)+'.csv', delimiter = ';')
    tmp_df['dia'] = i
    list_dfs.append(tmp_df)
    #print(len(tmp_df))
    
df =  pd.concat(list_dfs)
print(len(df))
```

```python
bcn_df = df[df['Estación'] ==  "Barcelona"]
print(len(bcn_df))
bcn_df.dropna()
print(len(bcn_df))
bcn_df
```

```python
bcn_df['Tmax'] = bcn_df["Temperatura máxima (ºC)"].apply(lambda x: float(str(x).split(' ')[0]))
bcn_df['Tmed'] = bcn_df["Temperatura media (ºC)"].apply(lambda x: float(str(x).split(' ')[0]))
```

```python
bcn_df['Tmax']
```

```python
import matplotlib.pyplot as plt
```

```python
fig, ax= plt.subplots()

bcn_df.plot(kind='line', x= 'dia', y='Tmed', ax=ax)
bcn_df.plot(kind='line', x= 'dia', y='Tmax', color= 'green', ax=ax)
```
